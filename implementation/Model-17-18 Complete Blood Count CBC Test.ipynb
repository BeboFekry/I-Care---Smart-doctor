```python
# For LLM
from langchain_google_genai import GoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from IPython.display import display, HTML
import markdown

# For OCR
import pytesseract
import cv2
```

```python
prompt_template = """you are a helpfull AI assistant that helps in giving information about Complete Blood Count (CBC) Test according to resluts what to eat, what to do, ...

"Given the CBC results you need": {information}
"""

# google_api_key = "AIzaSyBIvw7QEbrnN7HJTBqxu6CI_r7egCWf5tU"
google_api_key = 'AIzaSyCaPzDUJujNjXa8r2wQ5P0RCHlTMXJc5zE'
llm = GoogleGenerativeAI(model="gemini-1.5-flash", google_api_key=google_api_key, temprature=0)

prompt = PromptTemplate(input_variable=['information'],
                        template=prompt_template)
chain = LLMChain(llm=llm, prompt=prompt)
```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

def ocr(path):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    text = pytesseract.image_to_string(gray, lang='ara+eng')
    return text

def cbc(path, verbose=0):
    # get text from image
    text = ocr(path)
    response = chain.run(information=text)
    
    if verbose==1:
        response2 = markdown.markdown(response)
        display(HTML(response2))

    return response
```
